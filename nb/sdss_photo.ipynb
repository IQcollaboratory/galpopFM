Trying to settle a debate on whether to use the VAGC absmags in Jeremy's catalog or updated NASA-Sloan Atlas absmag. Seems VAGC absmags wins out.

```python
import os
import h5py
import numpy as np 
# -- plotting --
import matplotlib as mpl
import matplotlib.pyplot as plt 
mpl.rcParams['text.usetex'] = True
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['axes.linewidth'] = 1.5
mpl.rcParams['axes.xmargin'] = 1
mpl.rcParams['xtick.labelsize'] = 'x-large'
mpl.rcParams['xtick.major.size'] = 5
mpl.rcParams['xtick.major.width'] = 1.5
mpl.rcParams['ytick.labelsize'] = 'x-large'
mpl.rcParams['ytick.major.size'] = 5a
mpl.rcParams['ytick.major.width'] = 1.5
mpl.rcParams['legend.frameon'] = False
```

```python
dat_dir = os.environ['GALPOPFM_DIR']
```

```python
fsdss = os.path.join(dat_dir, 'obs', 'tinker_SDSS_centrals_M9.7.valueadd.hdf5') 
sdss = h5py.File(fsdss, 'r') 

z_sdss = sdss['redshift_tinker'][...]/299792.458

#FNugriz
Fmag    = sdss['ABSMAG'][...][:,0]
Nmag    = sdss['ABSMAG'][...][:,1]
Gmag    = sdss['ABSMAG'][...][:,3]
Rmag    = sdss['ABSMAG'][...][:,4]

_Rmag    = sdss['mr_tinker'][...]
_Gmag    = sdss['mg_tinker'][...]
```

```python
plt.scatter(Rmag, _Rmag)
plt.plot([-25, -10], [-25, -10], c='k', ls='--')
plt.plot([-25, -10], [-25-5.*np.log10(0.7), -10-5.*np.log10(0.7)], c='k', ls=':')

plt.xlabel('NSA $M_r$', fontsize=20)
plt.ylabel('Tinker VAGC $M_r - 5\log 0.7$', fontsize=20)
plt.xlim(-25, -16)
plt.ylim(-25, -16)
```

```python
plt.scatter(Gmag, _Gmag)
plt.plot([-25, -10], [-25, -10], c='k', ls='--')
plt.plot([-25, -10], [-25-5.*np.log10(0.7), -10-5.*np.log10(0.7)], c='k', ls=':')

plt.xlabel('NSA $M_g$', fontsize=20)
plt.ylabel('Tinker VAGC $M_g - 5\log 0.7$', fontsize=20)
plt.xlim(-25, -16)
plt.ylim(-25, -16)
```

```python
plt.scatter(Gmag - Rmag, _Gmag - _Rmag)
plt.plot([-0.5, 2], [-0.5, 2], c='k', ls='--')

plt.xlabel('NSA $G-R$', fontsize=20)
plt.ylabel('Tinker VAGC $G-R$', fontsize=20)
plt.xlim(-0.5, 2)
plt.ylim(-0.5, 2)
```
